## Imputation Missing Value

```python
data.isnull().sum() #nyari missing per variabel
```

```python
data.isnull().sum().sum() #ngitung jumlah semuanya bukan per var
```

```python
#imputasi missing value data numerik #sintaxnya sama terserah mau ngisi mean apa mediannya
data['X2'] = data['X2'].fillna((data['X2'].mean())) 
data['X4'] = data['X4'].fillna((data['X4'].mean())) 
data['X5'] = data['X5'].fillna((data['X5'].mean())) 
data['X6'] = data['X6'].fillna((data['X6'].mean())) 
data['X8'] = data['X8'].fillna((data['X8'].mean())) 
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Sex, Steroid, Anaroxia, Spleen Pabable, Spiders tanpa membuat kolom baru
data['Class'] = LabelEncoder().fit_transform(data['Class'])
data['X1'] = LabelEncoder().fit_transform(data['X1'])
data['X2'] = LabelEncoder().fit_transform(data['X2'])
data['X3'] = LabelEncoder().fit_transform(data['X3'])
data['X4'] = LabelEncoder().fit_transform(data['X4'])
data['X5'] = LabelEncoder().fit_transform(data['X5'])
data['X6'] = LabelEncoder().fit_transform(data['X6'])
data['X7'] = LabelEncoder().fit_transform(data['X7'])
data['X8'] = LabelEncoder().fit_transform(data['X8'])
data['X9'] = LabelEncoder().fit_transform(data['X9'])
data['X10'] = LabelEncoder().fit_transform(data['X10'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['X2'] = StandardScaler().fit(data[['X2']]).transform(data[['X2']])
data['X4'] = StandardScaler().fit(data[['X4']]).transform(data[['X4']])
data['X5'] = StandardScaler().fit(data[['X5']]).transform(data[['X5']])
data['X6'] = StandardScaler().fit(data[['X6']]).transform(data[['X6']])
data['X7'] = StandardScaler().fit(data[['X7']]).transform(data[['X7']])
data['X8'] = StandardScaler().fit(data[['X8']]).transform(data[['X8']])
data['X9'] = StandardScaler().fit(data[['X9']]).transform(data[['X9']])
data['X10'] = StandardScaler().fit(data[['X10']]).transform(data[['X10']])
data.head()
```
