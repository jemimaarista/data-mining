# Gabungan Lengkap

```python
import pandas as pd
import numpy as np
```

```python
## Merging
kereta = pd.DataFrame({'Kode Kereta': ['kode0', 'kode1', 'kode2', 'kode3'],
                     'Jenis Kereta': ['Ekonomi', 'Ekonomi', 'Bisnis', 'Bisnis'], 
                     'Jumlah Keberangkatan': ['1', '3', '2', '4']})

tiket = pd.DataFrame({'Kode Kereta': ['kode0', 'kode1', 'kode2', 'kode3'],
                      'Harga Tiket': ['150000', '100000', '125000', '250000'],
                      'Kota Asal': ['Bandung', 'Surabaya', 'Surabaya', 'Jakarta']})

result1 = pd.merge(kereta, tiket, on='Kode Kereta')
result1
```

```python
## Merging different key id
biodata = pd.DataFrame({'Kode Kereta': ['kode0', 'kode1', 'kode2', 'kode3','kode4','kode5'],
                     'Jenis Kereta': ['Ekonomi', 'Ekonomi', 'Bisnis', 'Bisnis','Bisnis','Ekonomi'], 
                     'Jumlah Keberangkatan': ['1', '3', '2', '4','5','2']})

nilai = pd.DataFrame({'Kode Kereta': ['kode0', 'kode1', 'kode2', 'kode3'],
                      'Harga Tiket': ['150000', '100000', '125000', '250000'],
                      'Kota Asal': ['Bandung', 'Surabaya', 'Surabaya', 'Jakarta']})

result2 = pd.merge(kereta, nilai, on='Kode Kereta')
result2
```

```python
# Ordering variable/column
result1 = result1[['Kode Kereta', 'Jenis Kereta', 'Jumlah Keberangkatan', 'Kota Asal','Harga Tiket']]
result1
```

```python
# Multiple key
kereta = pd.DataFrame({'Kode Kereta': ['kode0', 'kode0', 'kode1', 'kode1'],
                        'Kota Tujuan': ['Kediri', 'Banyuwangi', 'Kediri', 'Jogja'],
                       'Jenis Kereta': ['Ekonomi', 'Ekonomi', 'Bisnis', 'Bisnis'], 
                     'Jumlah Keberangkatan': ['1', '3', '2', '4']})

tiket = pd.DataFrame({'Kode Kereta': ['kode0', 'kode0', 'kode1', 'kode1'],
                      'Kota Tujuan': ['Kediri', 'Banyuwangi', 'Malang', 'Jogja'],
                      'Harga Tiket': ['150000', '100000', '125000', '250000'],
                      'Kota Asal': ['Bandung', 'Surabaya', 'Surabaya', 'Jakarta']})
result3 = pd.merge(kereta, tiket, on=['Kode Kereta', 'Kota Tujuan'])
result3
```

```python
# Merge methods (default how="inner")
result_left = pd.merge(kereta, tiket, how='left', on=['Kode Kereta', 'Kota Tujuan'])
result_left
```

```python
# Merge methods (default how="inner")
result_right = pd.merge(kereta, tiket, how='right', on=['Kode Kereta', 'Kota Tujuan'])
result_right
```

```python
result_outer = pd.merge(kereta, tiket, how='outer', on=['Kode Kereta', 'Kota Tujuan'])
result_outer
```

```python
result_inner = pd.merge(kereta, tiket, how='inner', on=['Kode Kereta', 'Kota Tujuan'])
result_inner
```

```python
## Joining 
kereta = pd.DataFrame({'Jenis Kereta': ['Ekonomi', 'Ekonomi', 'Bisnis'], 
                      'Jumlah Keberangkatan': ['1', '3', '2']},
                    index=['kode0', 'kode1', 'kode2'])
tiket = pd.DataFrame({'Kota Tujuan': ['Kediri', 'Banyuwangi', 'Malang'],
                      'Harga Tiket': ['150000', '100000', '125000']},
                    index=['kode0', 'kode2', 'kode3'])
result = biodata.join(tiket) #Based Kereta ditambahkan tiket
result
```

```python
result = nilai.join(kereta, how='outer') #Seluruh data masuk
result
```

```python
result = nilai.join(kereta, how='inner') # Hanya data yang lengkap
result
```

```python
#The data alignment here is on the indexes (row labels). 
#This same behavior can be achieved using merge plus additional arguments instructing it to use the indexes
result = pd.merge(kereta, tiket, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(kereta, tiket, left_index=True, right_index=True, how='inner')
result
```

```python
kereta = pd.DataFrame({'Jenis Kereta': ['Ekonomi', 'Bisnis', 'Ekonomi', 'Bisnis'], 
                      'Jumlah Keberangkatan': ['1', '3', '2', '4'],
                   'key': ['kd0', 'kd1', 'kd0', 'kd1']})
tiket = pd.DataFrame({'Kota Asal': ['Bandung', 'Jakarta'],
                       'Harga Tiket': ['100000', '210000']},
                      index=['kd0', 'kd1'])
result = kereta.join(tiket, on='key')
result
```

```python
## Concatenating
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

```python
# Add category
result = pd.concat(frames, keys=['X', 'Y', 'Z'])
result
```

```python
# Default join=outer (join all rows)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=0) #axis=1 is for concated by y-axis; default axis=0
result
```

```python
# Join = inner (join only rows that exist in both datasets) 
# hanya yang ada di df1 dan df4
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
# Join the exact rows in df1 
#Menggabungkan berdasarkan df1
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
## Concatenating using append
result = df1.append(df2) #default axis=0
result
```

```python
result = df1.append(df4)
result
```

```python
result = df1.append([df2, df3]) #multiple datasets to concatenate
result
```

```python
# For DataFrames which don’t have a meaningful index, 
#you may wish to append them and ignore the fact that they may have overlapping indexes.
result = pd.concat([df1, df4], ignore_index=True)
result
```

```python
result = df1.append(df4, ignore_index=True)
result
```

```python
# Appending rows to a DataFrame
s2 = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
result = df1.append(s2, ignore_index=True)
result
```

```python
dicts = [{'A': 1, 'B': 2, 'C': 3, 'X': 4}, #dictionary
         {'A': 5, 'B': 6, 'C': 7, 'Y': 8}]
result = df1.append(dicts, ignore_index=True)
result
```
