# Assosiation Rule

```python
import numpy as np  
import pandas as pd  
from apyori import apriori
```

```python
#Membaca Data
dataor=pd.read_csv('online_retail.csv')
dataor.head()
```

```python
#Mengetahui Missing Value
dataor.isnull().sum()
```

```python
#Imputasi Missing Value Variabel "Description"
dataor['Description'] = dataor['Description'].str.strip()
dataor.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
#Imputasi Missing Value Variabel "Invoiceno" Awalan C
dataor['InvoiceNo'] = dataor['InvoiceNo'].astype('str')
dataor = dataor[~dataor['InvoiceNo'].str.contains('C')]
```

```python
#Imputasi Ulang Missing Value Variabel "Description"
dataor.dropna(axis=0, subset=['Description'], inplace=True)
```

```python
#Imputasi Ulang Missing Value Variabel "Customer ID"
dataor.dropna(axis=0, subset=['CustomerID'], inplace=True)
```

```python
#Imputasi Missing Value Variabel "Country"
dataor.dropna(axis=0, subset=['Country'], inplace=True)
```

```python
#Cek Ulang Missing Value
dataor.isnull().sum()
```

```python
#Pengelompokan Transaksi Berdasarkan Variabel "InvoiveNo"
y = dataor.groupby('InvoiceNo')
transaksi = []
for i,j in y:
    transaksi.append(list(j['Description'].map(str)))
```

```python
#Tabel Transaksi
datatrans = pd.DataFrame(transaksi)
datatrans.head()
```

```python
#Cek Jumlah Kolom dan Baris
datatrans.shape
```

```python
records=[]
for i in range (0, 18536):
    records.append([str(datatrans.values[i,j]) for j in range (0,542)])
```

```python
association_rules=apriori(records, min_support=0.005,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
#Index Pertama Untuk Inner List
#Base Items & Add Items
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items [0] + " -> " + items [1])
    
    #Index Kedua Untuk Inner List
    print("Support: " + str(item[1]))
    
    #Index Ketiga Untuk List ke-10
    #Index Ketiga Inner List
    
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
